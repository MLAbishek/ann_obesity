```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df=pd.read_csv(r'D:\programming\coding\dataset\newobs.csv')
df.head()
print(df['NObeyesdad'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
df.loc[:,'Gender']=lr.fit_transform(df.loc[:,'Gender'])
df.loc[:,'family_history_with_overweight']=lr.fit_transform(df.loc[:,'family_history_with_overweight'])
df.loc[:,'FAVC']=lr.fit_transform(df.loc[:,'FAVC'])
df.loc[:,'CAEC']=lr.fit_transform(df.loc[:,'CAEC'])
df.loc[:,'SMOKE']=lr.fit_transform(df.loc[:,'SMOKE'])
df.loc[:,'SCC']=lr.fit_transform(df.loc[:,'SCC'])
df.loc[:,'CALC']=lr.fit_transform(df.loc[:,'CALC'])
df.loc[:,'MTRANS']=lr.fit_transform(df.loc[:,"MTRANS"])
df.loc[:,'NObeyesdad']=lr.fit_transform(df.loc[:,'NObeyesdad'])
df['NObeyesdad']=df['NObeyesdad'].astype('float64')
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
print(y_train)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=17,activation='relu'))
ann.add(tf.keras.layers.Dense(units=22,activation='relu'))
ann.add(tf.keras.layers.Dense(units=7,activation='softmax'))
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
ann.fit(x_train,y_train,batch_size=32,epochs=150)
```

```python
l=ann.predict(x_train,y_train)
y_pred=np.argmax(l,axis=1)
print(y_pred)
```
